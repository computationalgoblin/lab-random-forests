```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from scipy.stats import norm
from sklearn.preprocessing import OneHotEncoder
```

```python
numerical_data = pd.read_csv(r'files_for_lab\numerical.csv')
categorical_data = pd.read_csv(r'files_for_lab\categorical.csv')
target = pd.read_csv(r'files_for_lab\target.csv')
```

```python
display(numerical_data.head())
display(categorical_data.head())
display(target.head())
```

Preprocessing the data

```python
encoder = OneHotEncoder(drop='first', sparse=False)
encoder.fit(categorical_data)
categorical_encoded = encoder.transform(categorical_data)
categorical = pd.DataFrame(categorical_encoded)

data = pd.concat([numerical_data, categorical, target], axis=1)
```

Train_test_splitting

```python
X = data.drop('TARGET_B', axis=1)
y = data['TARGET_B']

X.columns = X.columns.astype(str)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train, y_train)
```

Random Forest Classifier

```python
rfc = RandomForestClassifier(max_depth=2, random_state=0)
rfc.fit(X_train, y_train)
pred = rfc.predict(X_test)


print(classification_report(y_test, pred))
```
